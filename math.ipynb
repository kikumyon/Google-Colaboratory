# 方程式を解く

```python
from sympy import*
```

```python
x,y = symbols("x y")
```

**方程式を定義する**

Eqは「=」の意味である

```python
Eq(x, y)
```

```python
Eq(x **2,1)
```

**solve関数を用いて計算する**

```python
solve(Eq(x **2,1), x)
```

答えはリストで帰ってくる

一番後ろのxは「xについて解く」という意味である

**別解**

```python
Eq(x **2-1,0)
```

```python
solve(Eq(x **2-1,0), x)
```

**省略版**

```python
x **2-1
```

```python
solve((x **2-1), x)
```

# 連立方程式を解く

```python
Eq(4*x-3*y, -9)
```

```python
Eq(3*x-7*y,17)
```

**変数に置き換える**

```python
eq1=Eq(4*x-3*y, -9)
```

```python
eq2=Eq(3*x-7*y,17)
```

方程式が複数ある場合はリストで記す
解く文字が複数ある場合もリストで記す

```python
solve([eq1, eq2],[x, y])
```

# 積分

```python
exp(x)*sin(x)+exp(x)*cos(x)
```

積分する

```python
integrate(exp(x)*sin(x)+exp(x)*cos(x), x)
```

```python
sin(x**2)
```

sin(x**2)をxについて-∞〜∞の間で積分する

```python
integrate(sin(x**2), (x, -oo, oo))
```
